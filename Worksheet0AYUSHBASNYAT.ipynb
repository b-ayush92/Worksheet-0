```python
# Write a recursive function sum_nested_list(nested_list) that:
# 1. Takes a nested list (a list that can contain numbers or other lists of numbers) as
# input.
# 2. Sums all numbers at every depth level of the list, regardless of how deeply nested
# the numbers are.

def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if isinstance(element, list):
            # If the element is a list, recursively sum the nested list
            total += sum_nested_list(element)
        else:
            # If the element is a number, add it to the total
            total += element
    return total

# Test the function with a sample nested list
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)
print(f"The sum of all numbers in the nested list is: {result}")
    
```

```python
# • Write a recursive function generate_permutations(s) that:
# – Takes a string s as input and returns a list of all unique permutations.
# • Test with strings like ”abc” and ”aab”.

# print(generate_permutations("abc"))
# Should return [’abc’, ’acb’, ’bac’, ’bca’, ’cab’, ’cba’]

def generate_permutations(s):
    if len(s) == 1:
        return [s]
    
    permutations = []
    for i, char in enumerate(s):
        # Recursively generate permutations for the remaining characters
        for perm in generate_permutations(s[:i] + s[i+1:]):
            new_perm = char + perm
            if new_perm not in permutations:
                permutations.append(new_perm)
    
    return permutations
    
print(generate_permutations("abc"))

print(generate_permutations("aab"))
```

```python
# Write a recursive function calculate_directory_size(directory) where:
#  • directory is a dictionary where keys represent file names (with values as sizes in
#  KB) or directory names (with values as another dictionary representing a subdi
# rectory).
#  • The function should return the total size of the directory, including all nested
#  subdirectories.
#  2. Test the function with a sample directory structure.

def calculate_directory_size(directory):
    total_size = 0
    for key, value in directory.items():
        if isinstance(value, dict):
            total_size += calculate_directory_size(value)
        else:
            # If the value is a number, add it to the total size
            total_size += value
    return total_size

directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

result = calculate_directory_size(directory_structure)
print(f"The total size of the directory is: {result} KB")
```

```python
# 1. Write a function min_coins(coins, amount) that:
#  • Uses DP to calculate the minimum number of coins needed to make up the
#  amount.
#  2. Test with coins = [1, 2, 5] and amount = 11. The result should be 3 (using coins
#  [5, 5, 1]).

def min_coins(coins, amount):
    # Initialize a list to store the minimum number of coins for each amount from 0 to amount
    dp = [float('inf')] * (amount + 1)
    dp[0] = 0  # Base case: 0 coins are needed to make the amount 0

    for coin in coins:
        # Update the dp array for each amount from coin to amount
        for i in range(coin, amount + 1):
            dp[i] = min(dp[i], dp[i - coin] + 1)

    # If dp[amount] is still infinity, it means it's not possible to make the amount with the given coins
    return dp[amount] if dp[amount] != float('inf') else -1

coins = [1, 2, 5]
amount = 11
result = min_coins(coins, amount)
print(f"The minimum number of coins needed to make the amount {amount} is: {result}")
```

```python
# 1.Writeafunctionlongest_common_subsequence(s1,s2)that:
#  •UsesDPtofindthelengthoftheLCSoftwostringss1ands2.
 # 2. Testwithstringslike"abcde" and "ace";theLCSlengthshouldbe3 ("ace").

def longest_common_subsequence(s1, s2):
    # Create a 2D array to store the lengths of LCS
    dp = [[0] * (len(s2) + 1) for _ in range(len(s1) + 1)]

    for i in range(1, len(s1) + 1):
        for j in range(1, len(s2) + 1):
            if s1[i - 1] == s2[j - 1]:
                dp[i][j] = dp[i - 1][j - 1] + 1
            else:
                dp[i][j] = max(dp[i - 1][j], dp[i][j - 1])

    return dp[len(s1)][len(s2)]

s1 = "abcde"
s2 = "ace"
result = longest_common_subsequence(s1, s2)
print(f"The length of the LCS of '{s1}' and '{s2}' is: {result}")
```

```python
# 1. Write a function knapsack(weights, values, capacity) that:
#  • Uses DP to determine the maximum value that can be achieved within the given
#  weight capacity.
#  2. Test with weights [1, 3, 4, 5], values [1, 4, 5, 7], and capacity 7. The re
# sult should be 9.

def knapsack(weights, values, capacity):
    
    n = len(weights)
    dp = [[0 for _ in range(capacity + 1)] for _ in range(n + 1)]

    for i in range(1, n + 1):
        for w in range(1, capacity + 1):
            if weights[i - 1] <= w:
                dp[i][w] = max(dp[i - 1][w], dp[i - 1][w - weights[i - 1]] + values[i - 1])
            else:
                dp[i][w] = dp[i - 1][w]

    return dp[n][capacity]

# Test the function with the given example
weights = [1, 3, 4, 5]
values = [1, 4, 5, 7]
capacity = 7
result = knapsack(weights, values, capacity)
print(f"The maximum value that can be achieved within the weight capacity {capacity} is: {result}")
# Should return 9
```
